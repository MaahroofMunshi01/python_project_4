```python
# Data Analyzer and Transformer Program

data = []        # Global dataset
summary = {}     

def average(numbers):
    """Calculate average of numbers"""
    return sum(numbers) / len(numbers)



def show_values(*args):
    """Display multiple values using *args"""

    for i in args:
        print(i)



def show_summary(**kwargs):
    """Display dataset summary using **kwargs"""

    for key, value in kwargs.items():
        print(key, ":", value)



# Recursion Function
def factorial(n):
    """Calculate factorial using recursion"""

    if n == 0:
        return 1

    return n * factorial(n-1)



# Lambda Function
def filter_numbers(numbers, limit):
    """Filter values using lambda"""

    result = list(filter(lambda x:x >= limit, numbers))

    return result



# Return Multiple Values
def statistics(numbers):
    

    return min(numbers), max(numbers), sum(numbers), average(numbers)



# Global Keyword
def update_summary(numbers):
    """Update global summary"""

    global summary

    summary = {
        "Total Values": len(numbers),
        "Average": average(numbers)
    }



while True:


    print("""
    1. Enter Data
    2. Display Summary
    3. Factorial
    4. Filter Data
    5. Sort Data
    6. Statistics
    7. Exit
    """)


    choice = int(input("Enter choice: "))



    # Input Data
    if choice == 1:


        values = input(
            "Enter numbers separated by space: "
        )


        data = list(map(int, values.split()))


        update_summary(data)


        print("Data Stored")



    # Built-in Functions
    elif choice == 2:


        print("\nData Summary")

        print("Length:", len(data))

        print("Minimum:", min(data))

        print("Maximum:", max(data))

        print("Sum:", sum(data))


        show_summary(**summary)



    # Recursion
    elif choice == 3:


        num = int(input("Enter number: "))


        print(
            "Factorial:",
            factorial(num)
        )



    # Lambda
    elif choice == 4:


        limit = int(
            input("Enter threshold: ")
        )


        print(
            "Filtered Data:",
            filter_numbers(data, limit)
        )



    # Sorting
    elif choice == 5:


        print("1. Ascending")

        print("2. Descending")


        ch = int(input("Choice: "))


        if ch == 1:

            data.sort()

        else:

            data.sort(reverse=True)


        print("Sorted Data:", data)



    # Multiple Return
    elif choice == 6:


        minimum, maximum, total, avg = statistics(data)


        print("\nStatistics")

        print("Minimum:", minimum)

        print("Maximum:", maximum)

        print("Sum:", total)

        print("Average:", avg)



    elif choice == 7:


        print("Goodbye!")

        break



    else:

        print("Invalid Choice")
```
